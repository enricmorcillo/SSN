Comencem important les llibreries necessàries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definim els paràmetres del nostre beam

```python
# Beam parameters
l = 1.0  # Length (m)
w = 0.1  # Width (m)
t = 0.01  # Thickness (m)
E = 200e9  # Young's modulus (Pa)
I = (w * t**3) / 12  # Moment of inertia
F = 100  # Applied force (N)
```

Definim la nostra funció

```python
def beam_deflection(x):
    return (F * x**2 * (3*l - x)) / (6 * E * I)

# Compute deflection along the beam
x_vals = np.linspace(0, l, 100)
z_vals = beam_deflection(x_vals)
```

Gràfiquem els resultats

```python
# Plot deflection profile
plt.figure()
plt.plot(x_vals, z_vals, label='Deflection profile')
plt.xlabel('x (m)')
plt.ylabel('Deflection z (m)')
plt.title('Deflection of a Clamped-Free Beam')
plt.legend()
plt.grid()
plt.show()
```
